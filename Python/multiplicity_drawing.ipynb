```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot(measured, measured_err, restored, restored_err):
    xx = list(range(0, 10))
    plt.plot(xx, measured, color="black")
    plt.errorbar(xx, measured, yerr=measured_err, fmt="s", label='Measured', color="black")

    plt.plot(xx, restored, color="red", linestyle='dashed')
    plt.errorbar(xx, restored, yerr=restored_err, fmt="o", label='Restored', color="red")

    plt.legend()

    plt.xlabel('Number of Neutrons')
    plt.ylabel('Probability')
    plt.show()
```

```python
# Rf-260
file_name = '260Rf_restored.csv'
df = pd.read_csv(file_name)
measured = df["probability"].to_numpy()
measured_err = df["probability_unc"].to_numpy()
restored = df["restored"].to_numpy()
restored_err = df["restored_unc"].to_numpy()

plot(measured, measured_err, restored, restored_err)
```
